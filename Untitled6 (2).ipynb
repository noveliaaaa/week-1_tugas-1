```python
# 2. define features (X) and target (y)
# untuk memilih fitur dan target
```

```python
x = data[['Years of Experience']]  # features, menggunakan kolom yang ada
y = data['Salary']  # target, menggunakan kolom yang ada
```

```python
# 3. split data into train and test sets
# untuk membagi data menjadi train dan test set
```

```python
x_  , x_test, y_train, y_  = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# 4. create and train the linear regression model
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() # Menginisialisasi model Regresi Linier
model.fit(X_train, y_train)
```

```python
# 5. make predictions
```

```python
y_pred = model.predict(X_test)
```

```python
# 6. evaluate the model
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```

```python
plt.scatter(X_test, y_test, label="Age")
plt.scatter(X_test, y_pred, label="Salary")
plt.legend()
plt.title("Age vs Predicted Salary")
plt.show()
```
